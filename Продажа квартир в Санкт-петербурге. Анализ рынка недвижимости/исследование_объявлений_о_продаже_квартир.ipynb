### Обработка пропусков

Посчитаем пропуски.

```python
data.isna().sum()
```

Пропуски указаны в 14 столбцах и везде в разном количестве, кроме двух: `ponds_around3000`и `parks_around3000`. Можно проверить, есть ли в этом связь, но оставим это на последний шаг, а в начале изучим остальные 12. Пойдём по алфавиту - с `ceiling_height`.

9195 пропущенных значений - это почти 40% от общей выборки, удаление такого большого куска данных однозначно исказит анализ.
Необходимо заменить пропуски, но для начала посмотрим на объявления с пропущенными значениями в этом столбце, используя логическую индексацию и функцию `head()`:

```python
data.query('ceiling_height.isna()')
```

Во всех остальных стобцах этих объявлений есть значения, т.е. нельзя сказать что данные пропущены из-за системной ошибки для нескольких столбцов сразу. 
Природу этих пропусков можно объяснить как ошибкой при выгрузке, так и человеческим фактором(поскольку высота потолков не относится к картографическим данным, а значит, по условиям проекта эти данные вносились вручную). Возможно человек, заполняющий данные для анкеты просто не знал/поленился измерить высоту потолков, либо забыл её внести, или счёл эту информацию необязательной и тд.
Логичным источником значения для заполнения этих пропусков мне видятся медианные значения высоты для домов с равным количеством этажей. 

Сохраним список уникальных значений столбца `floors_total`, затем пустой список, который заполним медианами в цикле и заполним с их помощью пропуски в датасете:

```python
data['floors_total'].unique()
```

```python
#Сохраняем уникальные значения:
floors = [16., 11.,  5., 14., 12., 26., 24., 27.,  9.,  6., 17., 25.,  3.,
       23., 19., 18.,  4., 22.,  8.,  2., 10.,  7., 13., 15., 20., 28.,
       36., 21., 35., 34.,  1., 29., 52., 37.]

#Создаём пустой список для медиан:
medians = []


#Заполним список с медианами в цикле:
for i in range(len(floors)):
    
    #Создаём датафрейм с единственным числом этажей:
    df = data[data['floors_total']==floors[i]]
    
    #Считаем медиану:
    value = df['ceiling_height'].median()
    
    #Округляем до двух знаков после запятой:
    value = round(value, 2)
    
    #Добавляем в список:
    medians.append(value)

#Проверяем что получилось:
medians
```

Заполним пропуски при помощи цикла и логической индексации:

```python
for i in range(len(floors)):
    data.loc[data['floors_total']==floors[i], 'ceiling_height']=data.loc[data['floors_total']==floors[i], 'ceiling_height'].fillna(medians[i])

#проверим, остались ли пропуски:
data['ceiling_height'].isna().sum()
```

Посмотрим на строки с оставшимися пропусками:

```python
data[data['ceiling_height'].isna()]
```

Скорее всего, пропуски остались только в объявлениях, где также не заполнено количество этажей в доме. Проверим, так ли это:

```python
data.query('ceiling_height.isna() and not floors_total.isna()')
```

Всего одно значение с указанным количеством этажей. Видимо, больше таких домов в выборке не оказалось. Оставшиеся пропуски составляют менее половины процента от выборки и их можно удалить. Помним, что пропусков в столбце `floors_total` тоже немного - 86 и многие объявлений с этими пропущенными значениями также содержат пропуски в столбце с высотой потолков, поэтому можно удалить их разом:

```python
#Указываем в аргумент функции два столбца:
data = data.dropna(subset=['ceiling_height', 'floors_total'])

#Проверяем, что код сработал корректно:
data.isna().sum()
```

Код отработал корректно.

Можно двигаться дальше. Столбец `living_area` - выведем 20 значений:

```python
data[data['living_area'].isna()].head(20)
```

Как вариант заполнения пропусков в этом столбце может быть использовано медианное значение по квартирам с одинаковым количеством комнат.

Повторим алгоритм предыдущего заполнения:

```python
data['rooms'].unique()
```

```python
#копируем уникальные значения количества комнат:
rooms = [ 3,  1,  2,  4,  5,  6,  0,  7, 11,  8, 14, 10, 15,  9, 19, 16, 12]

#перезапишем переменную, чтобы не плодить слишком много переменных:
medians = []

#Пишем цикл для заполнения медианных значений:
for room in range(len(rooms)):
    
#Создаём датафрейм с единственным комнат в квартире:    
    df = data[data['rooms']==rooms[room]]
    
    #Считаем медиану:
    value = df['living_area'].median()
    
    #Округляем до двух знаков после запятой:
    value = round(value, 2)
    
    #Добавляем в список:
    medians.append(value)
    
#Проверяем:
medians
```

Заполним пропуски в столбце `living_area` медианными значениями при помощи циклов и логической индексации:

```python
for i in range(len(rooms)):
    data.loc[data['rooms']==rooms[i], 'living_area']=data.loc[data['rooms']==rooms[i], 'living_area'].fillna(medians[i])

#проверим, остались ли пропуски:
data['living_area'].isna().sum()
```

Код отработал корректно, следующий на очереди столбец `is_apartment` - 20838 пропусков. Здесь логичным будет предположить что пропуски в этом столбце по умолчанию означают, что квартира является живой площадью и пропуски уместно будет заменить значением `False`.

```python
#Перезапишем столбец:
data['is_apartment'] = data['is_apartment'].fillna(False)

#Посмотрим на пропуски:
data.isna().sum()
```

Перейдём к столбцу `kitchen_area` . Выведем на экран 20 строк с пропусками в этом столбце:

```python
data[data['kitchen_area'].isna()].head(20)
```

Можно было предположить, что все квартиры с пропущенными значениями площади кухни это квартиры-студии. Но наблюдаются строки с 5 и 7 комнатными квартирами. Здесь также можно заполнить пропуски медианным значением. Единственное, что следует учитывать - кухни в студиях обычно меньше по площади, поэтому я бы разделил датасет на две категории: студии и 'остальные' и заполнял бы пропуски медианами, рассчитанными с этих двух категорий.

Чтобы не заводить отдельные переменные, которые больше не понадобятся, сделаем всё одной строкой:

```python
#Заполняем пропуски медианой по студиям:
data.loc[data['is_apartment']==True, 'kitchen_area'] = data.loc[data['is_apartment']==True, 'kitchen_area'].fillna(data.loc[data['is_apartment']==True, 'kitchen_area'].median())

#Заполняем пропуски медианой по 'остальным' квартирам:
data.loc[data['is_apartment']==False, 'kitchen_area'] = data.loc[data['is_apartment']==False, 'kitchen_area'].fillna(data.loc[data['is_apartment']==False, 'kitchen_area'].median())

#Проверим пропуски по датасету:
data.isna().sum()
```

Код отработал корректно. Теперь можно перейти к столбцу с балконами. Будет логично последовать подсказке) и предположить что в квартирах с пропущенным значением в этом столбце попросту нет балконов. В этом случае правильным будет заменить пропуски нулём:

```python
data['balcony']=data['balcony'].fillna(0)
data['balcony'].isna().sum()
```

Код отработал корректно.
Перейдём к столбцу `locality_name`. 48 пропусков - количетсво, которое можно безболезненно удалить.

Перезапишем таблицу.

```python
data=data.dropna(subset=['locality_name'])

#Посмотрим на оставшиеся пропуски:
data.isna().sum()
```

Во всех оставшихся столбцах с пропусками, кроме `days_exposition`, значения по условиям проекта проставлялись автоматически, т.е. имеет место исключительно системная ошибка при выгрузке. Вариант для заполнения мне видится только один - медианные значения по выборке для каждого столбца. Чтобы ускорить процесс сделаем это также в цикле:

```python
#Создадим список с названиями столбцов:
columns = ['airports_nearest', 'cityCenters_nearest', 'parks_around3000', 'parks_nearest', 'ponds_around3000', 'ponds_nearest']

#Пишем цикл:
for column in columns:
    #считаем медиану по столбцу:
    median = data[column].median()
    
    #Подставляем медиану в пропущенные значения
    data[column] = data[column].fillna(median)

#Проверяем как отработал код:
print(data.isna().sum())
data
```

Остался последний столбец с пропусками - `days_exposition`. 3171 пропущенных значений - это больше 10% от выборки, а значит жертвовать таким большим пластом данных нежелательно. Посмотрим на строки с пропусками более внимательно:

```python
data[data['days_exposition'].isna()]
```

По куску из таблицы трудно что-то сказать. Можно предположить, что эти объявления до сих пор не сняты. Чтобы проверить эту гипотезу, посмотрим на взаимосвязь объявлений с пропущенными значениями и датой публикации объявления. Для этого сначала переведём столбец `first_day_exposition` в формат даты и времени:

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'])
```

Теперь посмотрим на значения столбца с помощью "describe()"

```python
data['first_day_exposition'].describe()
```

Посмотрим на распределение объявлений с пропусками во времени с  помощью гистограммы:

```python
data.query('days_exposition.isna()')['first_day_exposition'].hist(bins=20, figsize=(10,5))
```

Как видим, чем ближе к дате последнего на момент выгрузки объявления, тем больше пропусков. Предположение о том, что эти объявления ещё активны можно условно подтвердить. Для исследования цена объекта недвижимости является ключевым показателем и брать для него данные с активных объявлений некорректно - перед продажей цена, вероятно будет изменена. Квартиры, которые не продаются несколько лет также, предположительно, выставлены по стоимости, неадекватной рынку. Также важной метрикой является сколько дней объявление было размещено - здесь также корректных данных нет- можно только предполагать и додумывать. Поэтому, несмотря на значительный объём таких объявлений - более 13%, лучше их убрать.

```python
data = data.dropna(subset=['days_exposition'])
data.isna().sum()
```

Пропусков не осталось, код сработал верно.

Чтобы не делать это на глаз, напишем функцию, считающую квартили по столбцу и печатающую на экране границы нормальных значений. Затем в пределах этих границ построим срез из датасета. Я намеренно не пишу срез в цикле, чтобы посмотреть на эти границы глазами и оценить их с точки зрения здравого смысла. Также 'ус' может быть меньше минимального и больше максимального значений, при этом варианте автозамена также будет ошибкой.

```python
#Объявляем функцию:
def iqr(column):
    #Находим первый квартиль:
    q1 = data[column].quantile(0.25)
    
    #Находим третий квартиль:
    q3 = data[column].quantile(0.75)
    
    #Считаем межквартильный размах:
    iqr = q3 - q1 
    
    print('нижняя граница стандартных значений =', round((q1 - 1.5*iqr), 2))
    print('верхняя граница стандартных значений =', round((q3 + 1.5*iqr), 2))
    
#Проверим работу функции на изучаемом столбце:
iqr('total_area')
```

Нижняя граница оказалась меньше нуля, поэтому можно оставить минимальное значение в 12 м2, верхнюю границу можно взять по гистограмме - 125 м2, воспользуемся логической индексацией и перезапишем таблицу:
